```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

#Import danych z xlsx
currency=pd.read_excel(r"C:/NBP/currency_2012_2023.xlsx")

#zamiana na dataframe oraz ustawienie wyświetlania wszystkich kolumn
currency_pd=pd.DataFrame(currency)
pd.set_option('display.max_columns', None)

#zmiana nazwy poszczególnych kolumn
currency_pd.rename(columns={'w':'Data','1USDdolar amerykański':'USD','1EUReuro':'EUR','1RUBrubel rosyjski':'RUB'},inplace=True)                   

#konwersja z int do datetime
currency_pd['Data']=pd.to_datetime(currency_pd['Data'], format='%Y%m%d')
data_all=currency_pd[['Data','USD','EUR','RUB']]

#Przypisanie danych do poszczególnych zmiennych przedstawiające roczny kurs USD,EUR,RUB
data2012=data_all[data_all.Data.between('2012-1-1', '2012-12-31')]
data2013=data_all[data_all.Data.between('2013-1-1', '2013-12-31')]
data2014=data_all[data_all.Data.between('2014-1-1', '2014-12-31')]
data2015=data_all[data_all.Data.between('2015-1-1', '2015-12-31')]
data2016=data_all[data_all.Data.between('2016-1-1', '2016-12-31')]
data2017=data_all[data_all.Data.between('2017-1-1', '2017-12-31')]
data2018=data_all[data_all.Data.between('2018-1-1', '2018-12-31')]
data2019=data_all[data_all.Data.between('2019-1-1', '2019-12-31')]
data2020=data_all[data_all.Data.between('2020-1-1', '2020-12-31')]
data2021=data_all[data_all.Data.between('2021-1-1', '2021-12-31')]
data2022=data_all[data_all.Data.between('2022-1-1', '2022-12-31')]
data2023=data_all[data_all.Data.between('2023-1-1', '2023-12-31')]

##### $$$DOLAR$$$ #####

#Sredni roczny kurs USD
avgUSD2012=round(data2012['USD'].mean(),3)
avgUSD2013=round(data2013['USD'].mean(),3)
avgUSD2014=round(data2014['USD'].mean(),3)
avgUSD2015=round(data2015['USD'].mean(),3)
avgUSD2016=round(data2016['USD'].mean(),3)
avgUSD2017=round(data2017['USD'].mean(),3)
avgUSD2018=round(data2018['USD'].mean(),3)
avgUSD2019=round(data2019['USD'].mean(),3)
avgUSD2020=round(data2020['USD'].mean(),3)
avgUSD2021=round(data2021['USD'].mean(),3)
avgUSD2022=round(data2022['USD'].mean(),3)
avgUSD2023=round(data2023['USD'].mean(),3)

#min wartość dolara w poszczególnych latach
minUSD2012=round(data2012['USD'].min(),3)
minUSD2013=round(data2013['USD'].min(),3)
minUSD2014=round(data2014['USD'].min(),3)
minUSD2015=round(data2015['USD'].min(),3)
minUSD2016=round(data2016['USD'].min(),3)
minUSD2017=round(data2017['USD'].min(),3)
minUSD2018=round(data2018['USD'].min(),3)
minUSD2019=round(data2019['USD'].min(),3)
minUSD2020=round(data2020['USD'].min(),3)
minUSD2021=round(data2021['USD'].min(),3)
minUSD2022=round(data2022['USD'].min(),3)
minUSD2023=round(data2023['USD'].min(),3)


#max wartość dolara w poszczególnych latach
maxUSD2012=round(data2012['USD'].max(),3)
maxUSD2013=round(data2013['USD'].max(),3)
maxUSD2014=round(data2014['USD'].max(),3)
maxUSD2015=round(data2015['USD'].max(),3)
maxUSD2016=round(data2016['USD'].max(),3)
maxUSD2017=round(data2017['USD'].max(),3)
maxUSD2018=round(data2018['USD'].max(),3)
maxUSD2019=round(data2019['USD'].max(),3)
maxUSD2020=round(data2020['USD'].max(),3)
maxUSD2021=round(data2021['USD'].max(),3)
maxUSD2022=round(data2022['USD'].max(),3)
maxUSD2023=round(data2023['USD'].max(),3)

#max wartość dolara w poszczególnych latach
roznica2012=round((maxUSD2012-minUSD2012),3)
roznica2013=round((maxUSD2013-minUSD2013),3)
roznica2014=round((maxUSD2014-minUSD2014),3)
roznica2015=round((maxUSD2015-minUSD2015),3)
roznica2016=round((maxUSD2016-minUSD2016),3)
roznica2017=round((maxUSD2017-minUSD2017),3)
roznica2018=round((maxUSD2018-minUSD2018),3)
roznica2019=round((maxUSD2019-minUSD2019),3)
roznica2020=round((maxUSD2020-minUSD2020),3)
roznica2021=round((maxUSD2021-minUSD2021),3)
roznica2022=round((maxUSD2022-minUSD2022),3)
roznica2023=round((maxUSD2023-minUSD2023),3)







```

```python
 ### Wykres pudełkowy pozwalający sprawdzić rozkład danych ###

fig, axs = plt.subplots(2, 6)

# USD 2012
axs[0, 0].boxplot(data2012['USD'], 1)
axs[0, 0].set_title('USD 2012',weight='bold')

# USD 2013
axs[0, 1].boxplot(data2013['USD'], 1)
axs[0, 1].set_title('USD 2013',weight='bold')

# USD 2014
axs[0, 2].boxplot(data2014['USD'], 1)
axs[0, 2].set_title('USD 2014',weight='bold')

# USD 2015
axs[0, 3].boxplot(data2015['USD'], 1)
axs[0, 3].set_title('USD 2015',weight='bold')

# USD 2016
axs[0, 4].boxplot(data2016['USD'], 1)
axs[0, 4].set_title('USD 2016',weight='bold')

# USD 2017
axs[0, 5].boxplot(data2017['USD'], 1)
axs[0, 5].set_title('USD 2017',weight='bold')


# USD 2018
axs[1, 0].boxplot(data2018['USD'], 1)
axs[1, 0].set_title('USD 2018',weight='bold')

# USD 2019
axs[1, 1].boxplot(data2019['USD'], 1)
axs[1, 1].set_title('USD 2019'weight='bold')

# USD 2020
axs[1, 2].boxplot(data2020['USD'], 1)
axs[1, 2].set_title('USD 2020',weight='bold')

# USD 2021
axs[1, 3].boxplot(data2021['USD'], 1)
axs[1, 3].set_title('USD 2021',weight='bold')

# USD 2022
axs[1, 4].boxplot(data2022['USD'], 1)
axs[1, 4].set_title('USD 2022',weight='bold')

# USD 2023
axs[1, 5].boxplot(data2023['USD'], 1)
axs[1, 5].set_title('USD 2023',weight='bold')



fig.subplots_adjust(left=0.5, right=3.2, bottom=0, top=2,
                    hspace=0.2, wspace=0.5)

plt.show()
```

```python
### Wykres min max dolara ###

labels = ['2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020', '2021', '2022', '2023']
min_values = [minUSD2012, minUSD2013, minUSD2014, minUSD2015, minUSD2016, minUSD2017, minUSD2018, minUSD2019, minUSD2020, minUSD2021, minUSD2022, minUSD2023]
max_values = [maxUSD2012, maxUSD2013, maxUSD2014, maxUSD2015, maxUSD2016, maxUSD2017, maxUSD2018, maxUSD2019, maxUSD2020, maxUSD2021, maxUSD2022, maxUSD2023]

x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars
y=[1,2,3,4,5]

fig, ax = plt.subplots(figsize=(20,10))

rects1 = ax.bar(x - width/2, min_values, width, label='Min. wartość dolara')
rects2 = ax.bar(x + width/2, max_values, width, label='Maks. wartość dolara')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('$',fontsize=30)
ax.set_title('Minimalna oraz maksymalna wartość dolara w latach 2012-2023',fontsize=20,weight="bold")
ax.set_xticks(x, labels,fontsize=15,weight="bold")
ax.set_yticks(y)

ax.legend()

ax.bar_label(rects1, padding=3)
ax.bar_label(rects2, padding=3)

fig.tight_layout()

plt.show()
```

```python
### Wykres przedstawiający różnice pomiędzy maksymalna a minimalną wartością dolara ###

years=[2012,2013,2014,2015,2016,2017,2018,2019,2020,2021,2022,2023]
roznice=[roznica2012,roznica2013,roznica2014,roznica2015,roznica2016,roznica2017,roznica2018,roznica2019,roznica2020,roznica2021,roznica2022,roznica2023]
fig = plt.figure(figsize=(10, 5))
 
# creating the bar plot
plt.barh(years, roznice, color='maroon')
 
plt.xlabel("$",fontsize=20)
plt.ylabel("Rok",weight='bold',fontsize=15)
plt.title("Różnice pomiędzy minimalna wartością a maksymalna dolara w latach 2012-2023",weight='bold')
plt.show()
```

```python
### Wykres przedstawiający reakcję dolara na sytuację panującą na Ukrainie - rok 2022 ###

x1=data2022['USD']
y1=data2022['Data']
y1=pd.to_datetime(y1, format = '%Y-%m-%d %H:%M')
fig, ax = plt.subplots(figsize=(20,10))
ax.plot(y1,x1,color='red',linewidth=3,label='Linia wartości dolara')

ax.grid()
plt.title('Zachowanie waluty $ na sytuacje panującą na Ukrainie - rok 2022',fontsize=15,weight='bold')
plt.xlabel('Data',fontsize=15,weight='bold',labelpad=15)
plt.ylabel('$',fontsize=30,labelpad=15)
plt.axvline(pd.to_datetime('2022-02-24', format = '%Y-%m-%d'),label='Wybuch wojny na Ukrainie',color='blue',linewidth=3)
plt.legend()
plt.show()
```
